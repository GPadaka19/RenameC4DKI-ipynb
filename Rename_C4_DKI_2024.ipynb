```python
import csv

def read_csv(file_path, column_name):
    texts = []
    with open(file_path, newline='') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            texts.append(row[column_name])
    return texts

if __name__ == "__main__":
    csv_file_path = "TextToArray.csv"  # Ganti dengan path file CSV Anda
    column_name = "Nama-Web"  # Ganti dengan nama kolom yang berisi teks dalam file CSV Anda

    texts = read_csv(csv_file_path, column_name)

    print("Selamat datang di program Text to Array Generator!")
    print("Data teks dari file CSV telah dimuat.")

    print("\nOutput:")
    print("[")
    for i, text in enumerate(texts):
        if i != len(texts) - 1:
            print(f'  "{text}",')
        else:
            print(f'  "{text}"')
    print("]")
```
